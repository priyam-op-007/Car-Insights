```python
import pandas as pd
import matplotlib.pyplot as plt
# create a new column indicating whether the price is under 1000000
indian_cars.loc[indian_cars['Ex-Showroom_Price'] < 1000000, 'price_category'] = 'under 1000000'
# group the data by brand name and price category, and count the number of occurrences
grouped_data = indian_cars.groupby(['Make', 'price_category']).size().reset_index(name='count')
# filter the data for cars under 1000000
filtered_data = grouped_data[grouped_data['price_category'] == 'under 1000000'][['Make', 'count']]
# create a bar graph of the filtered data
filtered_data.plot.bar(x='Make', y='count')
# add labels and title to the graph
plt.xlabel("Brand")
plt.ylabel("Number of Cars")
plt.title("Number of Cars under Rs 10,00,000 by Brand")

# display the graph
plt.show()
```

```python
# create a new column indicating whether the price is above 1000000 and under 5000000
indian_cars.loc[(indian_cars['Ex-Showroom_Price'] > 1000000) & (indian_cars['Ex-Showroom_Price'] < 5000000), 'price_category'] = 'above 1000000 and under 5000000'

# group the data by brand name and price category, and count the number of occurrences
grouped_data = indian_cars.groupby(['Make', 'price_category']).size().reset_index(name='count')

# filter the data for cars above 1000000 and under 5000000
filtered_data = grouped_data[grouped_data['price_category'] == 'above 1000000 and under 5000000'][['Make', 'count']]

# create a bar graph of the filtered data
filtered_data.plot.bar(x='Make', y='count')

# add labels and title to the graph
plt.xlabel("Brand")
plt.ylabel("Number of Cars")
plt.title("Number of Cars above Rs 10,00,000 and under Rs 50,00,000 by Brand")

# display the graph
plt.show()
```
